# **Khám phá dữ liệu**

### Import thư viện

```python
import pandas as pd
from pandas.api.types import is_numeric_dtype

import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

### Import file

```python
df = pd.read_csv('./HR_Analytics.csv')
```

### Mỗi dòng có ý nghĩa gì? Có vấn đề các dòng có ý nghĩa khác nhau không?

### Mỗi cột có ý nghĩa gì?

```python
mean_col = pd.read_csv("columns_meaning.csv")
mean_col
```

### Mỗi cột hiện đang có kiểu dữ liệu gì? Có cột nào có kiểu dữ liệu chưa phù hợp để có thể xử lý tiếp hay không?

```python
type_col = {'field':[], 'type': []}

for col in df:
    type_col['field'].append(col)

    if is_numeric_dtype(df[col]):
        type_col['type'].append('numerical')
    else:
        type_col['type'].append('categorical')

pd.DataFrame(type_col)
```

### Với mỗi cột, các giá trị (dạng số, dạng phân loại) được phân bố như thế nào?

```python
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_cat = df.select_dtypes(include = ['object'])

df_num.hist(figsize=(18, 20))

fig, axes = plt.subplots(3, 3, figsize=(18, 10))
fig.suptitle('Categorical Distributions')
i = 0
for col in axes:
    for row in col:
        df_tmp = df_cat[df_cat.columns[i]]
        row.set_title(df_cat.columns[i])
        row.pie(df_tmp.value_counts(), 
              labels=df_tmp.value_counts().keys(),
              autopct='%1.1f%%',
              startangle=90,)
        i += 1
```

### Có cần phải tiền xử lý dữ liệu hay không và nếu có thì nhóm sinh viên cần phải xử lý như thế nào?

# Khám phá mối quan hệ trong dữ liệu

### Thu nhập trung bình hàng tháng theo trình độ học vấn

```python
edu_income = df.groupby('Education')['MonthlyIncome'].mean().reset_index()

plt.plot(edu_income['Education'], edu_income['MonthlyIncome'])
plt.xlabel('Trình độ học vấn', fontsize=20)
plt.ylabel('Thu nhập trung bình hàng tháng', fontsize=20)
plt.title('Thu nhập trung bình hàng tháng theo trình độ học vấn', fontsize=25)
plt.show()
```

* Trình độ học vấn tỉ lệ thuận với thu nhập hàng tháng

### Tỉ lệ nhân viên nghỉ việc theo phòng ban

```python
dep_att = df.groupby(['Department', 'Attrition'])['Attrition'].count().unstack()

dep_att['Attrition Rate'] = dep_att['Yes'] / (dep_att['Yes'] + dep_att['No'])

sns.barplot(x=dep_att.index, y=dep_att['Attrition Rate'])
plt.title('Tỉ lệ nhân viên nghỉ việc theo phòng ban', fontsize=15)
plt.xlabel('Phòng ban', fontsize=10)
plt.ylabel('Tỉ lệ nhân viên nghỉ việc', fontsize=10)
plt.ylim((0, 0.3))
for i, rate in enumerate(dep_att['Attrition Rate']):
    plt.text(i, rate+0.01, f'{rate:.2f} %', ha='center')
plt.show()
```

* Tỉ lệ nghỉ việc ở các phòng ban là tương đương nhau
=> Phòng ban không ảnh hưởng đến quyết định nghỉ việc

### Tỉ lệ nhân viên nghỉ việc theo overtime

```python
ovt_att = df.groupby(['OverTime', 'Attrition'])['Attrition'].count().unstack()

ovt_att['Attrition Rate'] = ovt_att['Yes'] / (ovt_att['Yes'] + ovt_att['No'])

sns.barplot(x=ovt_att.index, y=ovt_att['Attrition Rate'])
plt.title('Tỉ lệ nhân viên nghỉ việc theo overtime', fontsize=15)
plt.xlabel('OverTime', fontsize=10)
plt.ylabel('Tỉ lệ nhân viên nghỉ việc', fontsize=10)
plt.ylim((0, 0.5))
for i, rate in enumerate(ovt_att['Attrition Rate']):
    plt.text(i, rate+0.01, f'{rate:.2f} %', ha='center')
plt.show()
```

* Tỉ lệ nghỉ việc ở những nhân viên có overtime cao gấp 3 lần ở những nhân viên không overtime => overtime ảnh hưởng đến quyết định nghỉ việc

### Tỉ lệ nhân viên nghỉ việc theo cấp độ công việc

```python
joblvl_att = df.groupby(['JobLevel', 'Attrition'])['Attrition'].count().unstack()

joblvl_att['Attrition Rate'] = joblvl_att['Yes'] / (joblvl_att['Yes'] + joblvl_att['No'])

sns.barplot(x=joblvl_att.index, y=joblvl_att['Attrition Rate'])
plt.title('Tỉ lệ nhân viên nghỉ việc theo cấp độ công việc', fontsize=15)
plt.xlabel('Cấp độ công việc', fontsize=10)
plt.ylabel('Tỉ lệ nhân viên nghỉ việc', fontsize=10)
plt.ylim((0, 0.4))
for i, rate in enumerate(joblvl_att['Attrition Rate']):
    plt.text(i, rate+0.01, f'{rate:.2f} %', ha='center')
plt.show()
```

* Những nhân viên cấp độ thấp hơn có xu hướng nghỉ việc cao hơn => Cấp độ công việc có ảnh hưởng đến tỉ lệ nghỉ việc

### So sánh giữa nhân viên nghỉ việc và không

```python
male_emp = df[df['Attrition'] == 'No']
female_emp = df[df['Attrition'] == 'Yes']

male_med = male_emp.groupby("Age")[['MonthlyIncome']].median().reset_index()
female_med = female_emp.groupby("Age")[['MonthlyIncome']].median().reset_index()

plt.plot(male_med['Age'], male_med['MonthlyIncome'], label ='Nhân viên không nghỉ việc')
plt.plot(female_med['Age'], female_med['MonthlyIncome'], label ='Nhân viên nghỉ việc')

plt.title('So sánh thu nhập trung bình giữa nhân viên nghỉ việc và nhân viên ở lại theo tuổi', fontsize=15)
plt.xlabel("Tuổi", fontsize=10)
plt.ylabel("Thu nhập trung bình hàng tháng", fontsize=10)
plt.legend()
plt.show()
```

* Nhân viên không nghỉ việc có xu hướng nhận được thu nhập cao hơn ở cùng độ tuổi so với nhân viên nghỉ việc

### Liên hệ giữa số lượng người nghỉ việc, giới tính và chuyên ngành

```python
plt.figure(figsize = (10,5))

df_attrition = df.loc[df['Attrition'] == 'Yes']

plt.title("Số lượng người nghỉ việc theo giới tính và chuyên ngành")
sns.countplot(x = 'Gender', hue = 'EducationField', data = df_attrition)
```

* Ở ngành Life Sciences và Medical có số lượng nghỉ việc nhiều hơn so với các ngành khác ở cả nam và nữ

### Liên hệ giữa độ hài lòng với công việc, thời gian làm việc và phòng ban

```python
plt.figure(figsize = (10,5))

ax = sns.lineplot(data=df, x='YearsAtCompany', y='JobSatisfaction', estimator=np.mean,
                  hue='Department', errorbar=('ci', 0))

ax.set_xlabel('Thời gian làm việc ở công ty')
ax.set_ylabel('Mức độ hài lòng với công việc')
ax.set_title('Mức độ hài lòng với công việc theo thời gian làm việc tại từng phòng ban', fontsize=15)
```

* Phòng Research & Development: Ở khoảng thời gian 0-25 năm làm việc thì các nhân viên với mức độ hài lòng ổn định từ 2.5-3. Ở khoảng thời gian từ 25-40 năm thì độ hài lòng biến động nhiều hơn ở mức 1-4, đặc biệt biến động nhiều nhất ở các mốc 25, 30, 35.
* Phòng Sales: Ở khoảng 0-15 năm, mức độ hài lòng của các nhân viên dao động trong khoảng 2.5-3.75. Ở khoảng thời gian từ 15-40 năm thì độ hài lòng biến động nhiều hơn ở mức 1-4, đặc biệt biến động nhiều nhất ở các mốc 25, 35.
* Phòng Human Resources: Ở khoảng thời gian 0-10 năm mức độ hài lòng của các nhân viên ở múc 2.5-3.25. Ở khoảng thời gian 10-35 năm mức độ hài lòng dao động từ 1-4, biến động mạnh nhất ở mốc 20.
* => Trong 20 năm đầu làm việc, các nhân viên ở phòng Sales có độ hài lòng với công việc cao nhất, phòng Research & Development ổn định nhất. Từ 20-40 năm thì ở các phòng đều có sự biến động mạnh.

### Liên hệ giữa lương theo giờ, giới tính và số năm làm việc

```python
ax = sns.lineplot(data=df, x='YearsAtCompany', y='HourlyRate', estimator=np.mean,
                  hue='Gender', ci=0)

ax.set_xlabel('Số năm làm việc')
ax.set_ylabel('Lương theo giờ')
ax.set_title('Lương theo giờ dựa trên số năm làm việc và giới tính', fontsize=15);
```

* 0-10 năm làm việc: không có chênh lệch nhiều giữa nam và nữ.
* 10-15 năm làm việc: nhân viên nam có mức lương cao hơn nhẹ so với nữ.
* 15-20 năm làm việc: nhân viên nữ có mức lương cao hơn so với nam, đặc biệt ở các mốc 17, 24.
* 20-33 năm làm việc: nhân viên nam có mức lương chênh lệch mạnh so với nữ. Tuy nhiên múc lương ở nam bị giảm mạnh ở mốc 30.
* 33-40 năm làm việc: nhân viên nữ có mức lương tăng mạnh, cao hơn so với nam. Nhân viên nam bị giảm mạnh đặc biệt ở mốc 34.
